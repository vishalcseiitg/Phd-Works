```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gexf('sampled_APS_pacs052030.gexf')

# Adjust the parameters of the spring layout
pos = nx.spring_layout(G, iterations=15, seed=1721)

# Customize visualization settings
plt.figure(figsize=(15, 9))
nx.draw(G, pos=pos, with_labels=False, node_color='blue', node_size=15, edge_color='black', width=0.15)

plt.axis('off')  # Turn off the axis
plt.tight_layout()  # Adjust spacing

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gexf('sampled_APS_pacs052030.gexf')

degrees = dict(G.degree())
degree_values = list(degrees.values())

# Plot the degree distribution with enhanced features
fig, ax = plt.subplots(figsize=(8, 6))

counts, bins, patches = ax.hist(degree_values, bins=20, edgecolor='white', alpha=0.8)

ax.set_xlabel("Degree", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("Degree Distribution", fontsize=14)

# Add count labels to each bar
for count, patch in zip(counts, patches):
    if count > 0:
        x = patch.get_x() + patch.get_width() / 2
        y = patch.get_height()
        ax.annotate(str(int(count)), xy=(x, y), xytext=(0, 5),
                    textcoords="offset points", ha='center', va='bottom')

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gexf('sampled_APS_pacs052030.gexf')

clustering_coefficients = nx.clustering(G)

# Plot the clustering coefficient distribution with enhanced features
fig, ax = plt.subplots(figsize=(8, 6))

counts, bins, patches = ax.hist(list(clustering_coefficients.values()), bins=20, edgecolor='white', alpha=0.8)

ax.set_xlabel("Clustering Coefficient", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("Clustering Coefficient Distribution", fontsize=14)

# Add count labels to each bar
for count, patch in zip(counts, patches):
    if count > 0:
        x = patch.get_x() + patch.get_width() / 2
        y = patch.get_height()
        ax.annotate(str(int(count)), xy=(x, y), xytext=(0, 5),
                    textcoords="offset points", ha='center', va='bottom')

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gexf('sampled_APS_pacs052030.gexf')

# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Plot the degree centrality histogram with enhanced features
fig, ax = plt.subplots(figsize=(8, 6))

counts, bins, patches = ax.hist(list(degree_centrality.values()), bins=20, edgecolor='white', alpha=0.8)

ax.set_xlabel("Degree Centrality", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("Degree Centrality Distribution", fontsize=14)

# Add count labels to each bar
for count, patch in zip(counts, patches):
    if count > 0:
        x = patch.get_x() + patch.get_width() / 2
        y = patch.get_height()
        ax.annotate(str(int(count)), xy=(x, y), xytext=(0, 5),
                    textcoords="offset points", ha='center', va='bottom')

plt.show()
```
